```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

A. 데이터 전처리 - 데이터 읽어오기

```python
csv_path="NFLX.csv"
df= pd.read_csv(csv_path,usecols=range(5))
df
```

B. 데이터 전처리 – 2021년 이후의 종가 데이터 출력하기

```python
df['Date']=pd.to_datetime(df['Date']) # 자료형 날짜로 변경
df_after_2021=df[df['Date']>="2021-01-01"] # 2021년 이후의 데이터만 필터링
df_after_2021
```

```python
# Matplotlib를 이용해서 데이터 시각화
plt.plot(df_after_2021['Date'], df_after_2021['Close']) # x축에 Date, y축에 Close

plt.title('NFLX Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')

# x축 45도 회전
plt.xticks(rotation=45)

plt.show()
```

C. 데이터 분석 - 2021년 이후 최고, 최저 종가 출력하기

```python
max_price = df_after_2021['Close'].max()
min_price = df_after_2021['Close'].min()
print(f'최고 종가: {max_price}')
print(f'최저 종가: {min_price}')
```
